```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BinaryStep(x):
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, BinaryStep(x))
plt.axis('tight')
plt.title('Activation Function : BinaryStep')
plt.show()
```

```python
def Linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, Linear(x))
plt.axis('tight')
plt.title('Activation Function : Linear')
plt.show()
```

```python
def Sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, Sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')
plt.show()
```

```python
def Tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, Tanh(x))
plt.axis('tight')
plt.title('Activation Function : Tanh')
plt.show()
```

```python
def ReLu(x):
    x1 = []
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(1)
    return x1
            
```

```python
x = np.linspace(-10, 10)
plt.plot(x, ReLu(x))
plt.axis('tight')
plt.title('Activation Function : ReLu')
plt.show()
```

```python
def Softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, Softmax(x))
plt.axis('tight')
plt.title('Activation Function : Softmax')
plt.show()
```
